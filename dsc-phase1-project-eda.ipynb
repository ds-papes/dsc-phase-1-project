## Business Problem

Microsoft sees all the big companies creating original video content and they want to get in on the fun. They have decided to create a new movie studio, but they don’t know anything about creating movies. You are charged with exploring what types of films are currently doing the best at the box office. You must then translate those findings into actionable insights that the head of Microsoft's new movie studio can use to help decide what type of films to create.

```python
# Import standard packages
import os,glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Load up provided data files
folder = "zippedData/"
os.listdir(folder)
```

```python
f"{folder}*.csv*"
```

```python
# Preview files
files = glob.glob(f"{folder}*.gz*")
files
```

```python
# Transform filenames
files[0].replace('.csv.gz', '').split('/')[-1].replace('.', '_')
```

```python
# Load all .csv files into a dictionary using filename as key
tables = {}
dashes = '---'*25

for file in files:
    if file.endswith(".csv.gz"):
        # Save a variable-friendly version of filename
        table_name = file.replace('.csv.gz', '').split('/')[-1].replace('.', '_')
        print(dashes)

        # Load and preview dataframes
        print(f"Preview of {table_name}")
        tables[table_name] = pd.read_csv(file)
        display(tables[table_name].head(10))
        print()
    elif file.endswith(".tsv.gz"):
        # Save a variable-friendly version of filename
        table_name = file.replace('.tsv.gz', '').split('/')[-1].replace('.', '_')
        print(dashes)

        # Load and preview dataframes
        print(f"Preview of {table_name}")
        tables[table_name] = pd.read_csv(file, encoding='latin-1', delimiter='\t')
        display(tables[table_name].head(10))
        print()
```
